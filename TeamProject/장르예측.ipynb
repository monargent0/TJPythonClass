```python
# import
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

### 데이터 불러오기

```python
songs = pd.read_csv("songs_fin.csv")
songs.head()
```

### 장르별 샘플 수 확인

```python
songs['genre'].value_counts()
```

### 데이터 정제 None, <> , 불경 등

```python
songs[songs['lyric'] == 'None']['genre'].value_counts()
```

```python
print(559,
486 - 29,
448,
435,
433 - 1,
418 - 108)
```

```python
songs.drop(songs[songs['lyric'] == 'None'].index , inplace= True)
songs
```

```python
songs.drop(songs[songs['lyric'].str.contains("<")].index , inplace= True)
songs
```

```python
songs.drop(songs[songs['artist_name'].str.contains("스님")].index , inplace= True)
songs
```

```python
songs[songs['lyric'].str.contains('개리')]
```

### 데이터 백업

```python
songs = songs[['genre','song_name',"artist_name",'lyric']]
rnn = songs[['genre','lyric']]
```
